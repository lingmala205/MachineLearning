```python
from sklearn import preprocessing                      #Import preprocessing.
from sklearn.naive_bayes import GaussianNB             #Import Naive Bayes methods.


#Create a preprocessing enabler to convert strings into usable numbers.
numbers = preprocessing.LabelEncoder()

#Create a classifier to train the data.
model = GaussianNB()


#Away = 0, Home = 1
location = ['Home','Away','Home','Home','Away',
            'Home','Home','Away','Away','Home',
            'Away','Away','Away','Home','Home',
            'Home','Home','Away','Home','Home',
            'Home','Home','Home','Away']

#Convert the data.
locationNumbers = numbers.fit_transform(location)


#In = 0, Out = 1
opponent = ['Out','Out','In','Out','In',
            'Out','In','Out','Out','Out',
            'Out','In','Out','Out','Out',
            'Out','Out','Out','In','Out',
            'Out','Out','In','In']

#Convert the data.
opponentNumbers = numbers.fit_transform(opponent)


#Lose = 0, Win = 1
label =    ['Win','Win','Win','Win','Lose',
            'Win','Win','Win','Win','Win',
            'Win','Lose','Lose','Win','Lose',
            'Lose','Win','Lose','Lose','Win',
            'Lose','Win','Lose','Lose']

#Convert the data.
labelNumbers = numbers.fit_transform(label)

#Print the converted data to verify the conversions.
print(locationNumbers)
print(opponentNumbers)
print(labelNumbers)

#Leave space between output.
print(" ")
print(" ")
print(" ")
print(" ")

#Create a zip to include on the model for the testing data.
combine = list(zip(locationNumbers,opponentNumbers))

#Note: print(list(zip(locationNumbers,opponentNumbers))) also works.

#Print the zip variable to confirm assembly of data to be tested.
print(combine)

#Train the classification model by fitting the training data.
model.fit(combine,labelNumbers)


#For Id's 25 to 36
#.predict(location,opponent)
#location:        Away = 0, Home = 1
#opponent:        In   = 0, Out  = 1
#Predicted label: Lose = 0, Win  = 1 <------Output


#Make predictions using the data for the Win/Lose label
#for ID numbers 25 to 36 in the testing data.

#For Id's 25 to 36
# expectedValue = model.predict(location,opponent)
expectedValue25 = model.predict([[1,1]])
expectedValue26 = model.predict([[1,0]])
expectedValue27 = model.predict([[0,1]])
expectedValue28 = model.predict([[0,1]])
expectedValue29 = model.predict([[1,1]])
expectedValue30 = model.predict([[0,1]])
expectedValue31 = model.predict([[1,0]])
expectedValue32 = model.predict([[1,1]])
expectedValue33 = model.predict([[1,1]])
expectedValue34 = model.predict([[0,0]])
expectedValue35 = model.predict([[1,1]])
expectedValue36 = model.predict([[0,0]])


#Leave space between output.
print(" ")
print(" ")
print(" ")
print(" ")


#Print out the expected label results.
#Predicted label: Lose = [0], Win  = [1] <------Output
print('Predicted label: Lose = [0], Win  = [1]')
print(" ")
print("ID 25 Expected: ", expectedValue25)
print("ID 26 Expected: ", expectedValue26)
print("ID 27 Expected: ", expectedValue27)
print("ID 28 Expected: ", expectedValue28)
print("ID 29 Expected: ", expectedValue29)
print("ID 30 Expected: ", expectedValue30)
print("ID 31 Expected: ", expectedValue31)
print("ID 32 Expected: ", expectedValue32)
print("ID 33 Expected: ", expectedValue33)
print("ID 34 Expected: ", expectedValue34)
print("ID 35 Expected: ", expectedValue35)
print("ID 36 Expected: ", expectedValue36)
```
